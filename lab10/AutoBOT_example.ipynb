```python
import autoBOTLib
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
```

```python
# Load 20 Newsgroups dataset
newsgroups = fetch_20newsgroups(subset='all', categories=['rec.sport.baseball', 'sci.space'], remove=('headers', 'footers', 'quotes'))
```

```python
# Create a DataFrame for consistency
dataframe = pd.DataFrame({
    'text_a': newsgroups.data,
    'label': newsgroups.target
})
```

```python
dataframe.shape
```

```python
#Sample few percents for demonstration
dataframe = dataframe.sample(frac=0.1)
```

```python
dataframe.shape
```

```python
# Split into train and test sets
train_df, test_df = train_test_split(dataframe, test_size=0.2, random_state=42)
```

```python
# Prepare training data
train_sequences = train_df["text_a"].tolist()
train_targets = train_df["label"].tolist()
```

```python
# Initialize the object + conduct model search (cross-val is default evaluation)
autoBOTLibObj = autoBOTLib.GAlearner(
    train_sequences,
    train_targets,
    time_constraint=0.1
).evolve()
```

```python
# Summarize run
autoBOTLibObj.generate_report(output_folder="report", job_id="testJobId123")
```

```python
# Store model
autoBOTLib.store_autobot_model(autoBOTLibObj, "model.pickle")
```

```python
# Load an existing model
autoBOTObj = autoBOTLib.load_autobot_model("model.pickle")
```

```python
# Prepare test data
test_sequences = test_df["text_a"].tolist()
```

```python
# Predict on new data
predictions_non_prob = autoBOTObj.predict(test_sequences)
predictions_prob = autoBOTObj.predict_proba(test_sequences)
```

```python
# Display results
print("Predictions (non-probabilistic):", predictions_non_prob[:5])
print("Predictions (probabilistic):", predictions_prob[:5])
```
