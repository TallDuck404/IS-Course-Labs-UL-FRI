```python
md = pd.read_csv("movies.txt", sep=",")
md.describe()
```

```python
length_l = (md['length'] >= 100).sum()
length_h = (md['length'] <= 100).sum()

ratings_count = {
    "below": length_l,
    "above": length_h
}
print(length_l)
print(length_h)
sns.barplot( data=ratings_count)
plt.ylabel("Number of titles")
plt.show()
```
